```julia
using DelimitedFiles
using Plots
```

```julia
A = readdlm("profits.txt", ',')
```

```julia
plot(A[:,1], A[:,2], st = :scatter, label = "", xlabel = "Population (10,000)", ylabel = "Profits (10,000\$)")
```

```julia
y = A[:,2]
```

```julia
N = length(y)
```

```julia
X= [ones(N) A[:,1]]
```

```julia
#Normal equation
θ_best = inv(X' * X) * X' * y
```

```julia
#Use Gradient Descent algorithm to estimate θ
#J = lossFunc
```

```julia
function loss(X, y, θ)
    prediction = X * θ
    u = prediction - y
    (u' * u) / length(y)
end
```

```julia
J0 = loss(X, y, [0; 0])
```

```julia
J_best = loss(X, y, θ_best)
```

```julia
function bgd(X, y, α = 0.01, maxIterations = 1000) :: Array{Float64,2}
    N, D = size(X)
    θs= zeros(D, maxIterations)
    for t = 1: maxIterations - 1
        ∇ = X' * (X * θs[:, t] - y) / N
        θs[:, t + 1] = θs[:, t] - α * ∇
    end
    θs
end
```

```julia
θs = bgd(X, y, 0.01, 10)
```

```julia
T = 3000
θs = bgd(X, y, 0.01, T)
ts = collect(1: T)
Js = map(t -> loss(X, y, θs[:, t]), ts)
Js[end]
```

```julia
plot(ts, Js, xlabel = "Number of Iterations", ylabel = "Cost J", label = "")
```
