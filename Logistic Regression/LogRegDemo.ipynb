```julia
function loss(X::Array{Float64, 2}, y::Array{Int}, θ::Array{Float64})::Float64
    N = length(y)
    h = sigmoid(X * θ)
    l = y' * log.(h) + (1 .- y)' * log.(1 .- h)
    J = -l/N + 0.
end    
```

```julia
D= size(X, 2)
θ = zeros(D)
```

```julia
loss(X, y, θ)
```

```julia
function gradient(X, y, θ)::Array{Float64}
    N = length(y)
    h = sigmoid.(X * θ)
    X' * (h - y)/N .+ 0.
end
```

```julia
gradient(X, y, θ)
```

```julia
#batch gradient descent algorithm
function bgd(X, y, θ_0, α = 0.001, T = 100)
    θ = θ_0
    θs = []
    for t = 1 : T
        ∇ = gradient(X, y, θ)
        θ = θ - α * ∇
        push!(θs, θ)
    end
    θs
end
```

```julia
θ_0 = zeros(D)
α = 0.01
T= 1000
θs = bgd(X, y, θ_0, α, T)
```

```julia
Js = map(θ -> loss(X, y, θ), θs)
```

```julia
plot(1:T, Js)
```

```julia
α = 0.001
θs = bgd(X, y, θ_0, α, T)
```

```julia
Js = map(θ -> loss(X, y, θ), θs)
plot(1:T, Js)
```
